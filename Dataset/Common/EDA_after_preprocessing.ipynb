```python
# 패키지 불러오기
import pandas as pd
import seaborn as sns
from EDA_pack import DataVisualizer
```

```python
# seaborn의 내장 데이터셋인 'tips' 데이터 불러오기
#data = sns.load_dataset('tips')
data_x = pd.read_csv('Made_2024-03-26_16-49\Remove_outlier_data\Remove_outlier_X_train_data.csv')
print(data_x.info())
print(data_x.describe())
data_y = pd.read_csv('Made_2024-03-26_16-49\Remove_outlier_data\Remove_outlier_y_train_data.csv')
```

```python
# 전처리로 분리되어있던 독립변수, 종속변수 병합
data = pd.concat([data_x,data_y],axis=1)
```

```python
# DataVisualizer 클래스
visualizer = DataVisualizer(data)
```

```python
#1. 상관관계 히트맵
# 상관 관계 히트맵 그리기
visualizer.plot_correlation_heatmap()
```

```python
#2. 범주형 분포
# 범주형 변수의 분포 그리기
visualizer.plot_categorical_distribution('day')
visualizer.plot_categorical_distribution('sex')
```

```python
#3. 수치형 분포
# 수치형 변수의 분포 그리기
visualizer.plot_numerical_distribution('total_bill')
```

```python
#4. 박스플롯 
# 박스플롯 그리기
visualizer.box_plot('day','total_bill')
```

```python
#5. 박스플롯 & 히스토그램
# 박스플롯 & 히스토그램 플롯   
visualizer.box_his_plot('day','total_bill')
```

```python
#6. plotly 패키지 활용한 박스플롯 & 히스토그램
# plotly 패키지 활용한 박스플롯 & 히스토그램 플롯   
visualizer.plotly_box_his_plot('total_bill')
```
